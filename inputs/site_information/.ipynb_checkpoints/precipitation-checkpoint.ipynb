## Derive daily precipitation at each of the five sites across a Southern California Climate Gradient


**Methods**: Linear scaling of $\psi$ with precipitation



**Code Reference**: https://github.com/stevenallison/UCIClimateExperiment/blob/master/Precipitation.R\

**NOTE**: data in Glassman et al. (2019) are field transplanting-focused, which used data of limited, specific years. Instead, this simulation project, in theory, should use climate data of broader ranges, though in practice, with data availability and accessibiliyt limititations, those field experimentation-based data could be used.


**Data Source**:

- Desert: [data acess page](https://wrcc.dri.edu/cgi-bin/rawMAIN.pl?caucde);[Station Site](https://deepcanyon.ucnrs.org/weather-data/)
- Scrubland:https://wrcc.dri.edu/cgi-bin/rawMAIN.pl?caucbu (Burns Pinon Ridge Reserve California);  ~~https://deepcanyon.ucnrs.org/weather-data/~~(`per Glassman et al. 2019 PNAS`);
- Grassland:
- Pine-Oak: https://wrcc.dri.edu/cgi-bin/rawMAIN.pl?caucja (Jame Reserve, California)
- Subalpine: https://raws.dri.edu/cgi-bin/rawMAIN.pl?caCMSJ(Mt. San Jacinto California); State Park:https://www.parks.ca.gov/?page_id=636)

### Water potential ($\psi$) of the Grassland site at Loma Ridge, CA

derived from measrued fuel moisture (Allison and Goulden 2017 SBB)

```python
grassland = pd.read_csv('precipitation/grassland_water_potential.csv', index_col=0)
grassland.head()
```

```python
year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(grassland['Psi'],color='red',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

### Derive $\psi$ for the other four sites

by LINEARLY scaling the grassland site $\psi$ with **Total Annual Precipitation (TAP)** at each site.

```python
## AS PER Glassman et al. 2019, PNAS
TAP = pd.Series(data=[213.5, 428.4, 569.4, 1415.8, 1376.5],
                index=['desert', 'scrubland', 'grassland', 'pineoak', 'subalpine'])
TAP
```

```python
scaling_factor = TAP.grassland * (1./TAP)
scaling_factor
```

## Desert $\psi$

```python
desert_psi = grassland['Psi']*scaling_factor.desert

year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(desert_psi,color='red',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

## Scrubland $\psi$

```python
year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(grassland['Psi']*scaling_factor.scrubland,color='red',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

```python
year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(grassland['Psi']*scaling_factor.pineoak,color='red',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

## Subalpine $\psi$

```python
year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(grassland['Psi']*scaling_factor.subalpine,color='red',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

```python
year = 10
fig, ax1 = plt.subplots(1, 1,figsize=(18,3))

ax1.plot(desert_psi,color='red',linewidth=2,alpha=0.6)
ax1.plot(scrubland_psi,color='brown',linewidth=2,alpha=0.6)
ax1.plot(grassland_psi,color='orange',linewidth=2,alpha=0.6)
ax1.plot(pineoak_psi,color='green',linewidth=2,alpha=0.6)
ax1.plot(subalpine_psi,color='blue',linewidth=2,alpha=0.6)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

plt.show()
```

## Concatenate the 5 sites

```python
desert_psi    = grassland['Psi']*scaling_factor.desert
scrubland_psi = grassland['Psi']*scaling_factor.scrubland
grassland_psi = grassland['Psi']
pineoak_psi   = grassland['Psi']*scaling_factor.pineoak
subalpine_psi = grassland['Psi']*scaling_factor.subalpine

gradient_psi = pd.concat([desert_psi,scrubland_psi,grassland_psi,pineoak_psi,subalpine_psi],
                         axis=1,ignore_index=False,sort=False)
gradient_psi.columns=['desert', 'scrubland', 'grassland', 'pineoak', 'subalpine']
gradient_psi
```

```python
gradient_psi.plot(figsize=(18,3))
plt.show()
```

```python
gradient_psi.to_csv('precipitation/gradient_psi.csv')
```
